```python
import pandas as pd
from time import strftime

filename = input('Enter full file name: ')
df = pd.read_csv(filename)

df = df[df['Type'] == 'Sent']

rests = ['Shaikpet', 'Kondapur', 'Yousufguda', 'Gachibowli', 'Istahgrill Himayat Nagar', 'Madhapur']
sales = {}
consumption = {}
raw_materials = []
for rest in rests:
    temp_df = df[df['To outlet'] == rest]
    if (len(temp_df.index) != 0):
        start_date = pd.to_datetime(min(temp_df['Created at'])).strftime('%d-%m-%Y')
        end_date = pd.to_datetime(max(temp_df['Created at'])).strftime('%d-%m-%Y')
        sales[rest] = int(input('Enter sale for {} from {} to {} from Sales Sheet is: '.format(rest, start_date, end_date)))
        rest_consumption = {}
        for item in temp_df['Items']:
            entities = item.split('; ')
            for entity in entities:
                split = entity.split(' - ')
                item_name = split[0]
                if item_name not in raw_materials:
                    raw_materials.append(item_name)
                item_qty = float(split[1])
                if item_name in rest_consumption.keys():
                    rest_consumption[item_name] += item_qty
                else:
                    rest_consumption[item_name] = item_qty
        consumption[rest] = rest_consumption
    else:
        print('No sent report found for {} from {} to {}'.format(rest, start_date, end_date))
    

raw_materials_cost = {
    'UN COOKED CHICKEN': 250,
    'PANEER': 220,
    'BASE SAUCE': 94.05,
    'HUMMUS': 93,
    'GREEN CHILLY': 75,
    'RUMALI': 5.5,
    'KHUBUS': 35,
    'CHEESE POPPERS': 365,
    'FRENCH FRIES': 270,
    'EGGS': 5,
    'HONEY MUSTARD': 172,
    'MINT MAYO': 172,
    'FALAFEL': 30,
    'OIL TIN': 1410
}

print()
print('Cost of raw materials: {}'.format(raw_materials_cost))

food_cost = {}
for rest, rest_consumption in consumption.items():
    cost = 0
    for raw_material, qty in rest_consumption.items():
        cost += raw_materials_cost[raw_material]*qty
    food_cost[rest] = cost

food_cost_precentage = {}

for rest, cost in food_cost.items():
    if sales[rest] != 0:
        food_cost_precentage[rest] = round((cost*100)/sales[rest], 1)
    else:
        food_cost_precentage[rest] = 0

print()
print('=== RESULTS ===')
print("Outlet wise sales from {} to {}:".format(start_date, end_date))
for i,v in sales.items():
    print('{}: {}'.format(i,v))
print()
print("Outlet wise food cost from {} to {}:".format(start_date, end_date))
for i,v in food_cost.items():
    print('{}: {}'.format(i,v))
print()
print("Outlet Wise food cost percentage from {} to {}".format(start_date, end_date))
for i,v in food_cost_precentage.items():
    print('{}: {}%'.format(i,v))
```
